### Análise dos dados brutos da API
Este notebook realiza uma análise inicial dos dados brutos obtidos da API, incluindo a verificação de valores nulos e a estrutura geral do DataFrame.
Ele também prepara o terreno para a próxima etapa de transformação e limpeza dos dados.


```python

# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
     

# Ler os dados do arquivo CSV
df = pd.read_csv('./data/processed/dados_tratados.csv')

# Exibir as primeiras linhas para análise inicial
df.head(10)
```

```python
#Dimensão do DataFrame
df.shape
```

```python

# Estatísticas
df.describe()
```

```python

# Valores Nulos
df.isnull().sum()
```

```python
df['review_score'].replace(np.nan, 0, inplace=True)
df['review_score'] = df['review_score'].astype(int)
df['order_id'].isnull().sum()
df.dropna(subset=["order_id", "product_id"], inplace=True)
df.shape
```

```python

# Valores nulos em gráfico

ax = df.isnull().sum().plot(kind='bar', figsize=(10, 5))
plt.title('Valores Nulos por Coluna')
plt.xlabel('Colunas')
plt.ylabel('Número de Valores Nulos')

# Adiciona o valor no topo de cada barra
for p in ax.patches:
    ax.annotate(int(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.show()



```

```python
# Convertendo colunas de data para o formato datetime

df["order_approved_at"] = pd.to_datetime(df["order_approved_at"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["order_delivered_carrier_date"] = pd.to_datetime(df["order_delivered_carrier_date"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["review_creation_date"] = pd.to_datetime(df["review_creation_date"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df["review_answer_timestamp"] = pd.to_datetime(df["review_answer_timestamp"], errors="coerce").dt.strftime('%d/%m/%Y %H:%M:%S')
df.info()
```

```python
df.dtypes
```

```python
pd.set_option("display.float_format", "{:.1f}".format)

for column in df.columns:
    if df[column].isnull().sum() > 0:
        if df[column].dtype in ['float64', 'int64']:
            mean_value = round(df[column].mean(), 1)
            df[column].fillna(mean_value, inplace=True)
            df[column] = df[column].round(1)  # Arredonda para uma casa decimal
            print(f'Coluna {column}: Substituído nulos pela média {mean_value}')
        else:
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)
            print(f'Coluna {column}: Substituído nulos pela moda {mode_value}')
            
df.isnull().sum()

```
